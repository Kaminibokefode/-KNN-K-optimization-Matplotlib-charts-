```python
#K_opt_knn module has only method named K_optknn
    #this method takes the list of number of neighbors as input along with 

import K_opt_knn

k=[5,10,20,30,40]
Tr_Acc,Tr_Precision,Tr_recall,Ts_Acc,Ts_Precision,Ts_recall=K_opt_knn.knn_optimization(k,x_train2,y_train2,x_test,y_test)
```
